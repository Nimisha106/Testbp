```python
#test bp
#here it is changed code
import networkx as nx
import matplotlib.pyplot as plot
G=nx.MultiGraph()
G.add_nodes_from([1,2,3,4,5,6,7])
G.add_edges_from([(1,4),(1,6),(1,6),(2,3),(2,5),(2,6),(2,6),(2,6),(2,7),(3,5),(4,7),(5,7),(5,6),(6,7)])
print(G)
Ad=nx.adjacency_matrix(G)
print(Ad.todense())
G.number_of_nodes()
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos,node_size=300,node_color="orange")
nx.draw_networkx_labels(G,pos)
axi=plot.gca()
for e in G.edges:
    axi.annotate("",
                 xy=pos[e[0]],
                 xycoords='data',
                 xytext=pos[e[1]],
                 textcoords='data',
                 arrowprops=dict(arrowstyle="-",
                                 color="0.6",
                                 shrinkA=1,shrinkB=6,
                                 patchA=None,patchB=None,
                                 connectionstyle="arc3,rad=rrr".replace('rrr',str(0.3*e[2]),
                                ),
                                ),
                )
    
plot.savefig('part2_1(a).png')
plot.axis('off')
plot.show()

degree_G=G.degree()
print("The degrees of each node in the network are ", degree_G)
```
